```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from root_numpy import root2array, root2rec, tree2rec, array2root
from ROOT import TH1F, TFile
import nputils as nputils
```

```python
run = 150
subrun1 = 36
subrun2 = 50
data_basedir = '/Users/davidkaleko/Data/UBCommish/noise_files/'
fname = data_basedir + 'noise_run%03d_subrun%03d_%03d.root'%(run,subrun1,subrun2)
wf_df = pd.DataFrame(root2array(fname,'tree'))
```

```python
f = ROOT.TFile(fname,"READ")
```

```python
ch_mapping = 'mapping_for_jeremy.txt'
mapping_df = pd.DataFrame(np.loadtxt(ch_mapping, delimiter='\t',dtype='int'))
mapping_df.columns = ['larsoft_chno','crate','slot','femch']
```

```python
grp = wf_df.groupby(['crate', 'slot', 'femch'])
rms_of_rms = []
wfs_good = []
wfs_bad = []
plt.title("RMS of RMS over Events: Run %d %d-->%d"%(run,subrun1,subrun2))
plt.xlabel("Event number")
plt.ylabel("RMS of RMS")
for key, grp_df in grp:
    sr = grp_df['rms']
    rms = sr.std()
    if ( rms > 15 ) and not ( key[0] == 9 and key[1] == 16 and key[2] == 19 ):
        print key
        larsoft_num = int(mapping_df.query('crate == %d and slot == %d and femch == %d'%(key[0],key[1],key[2]))['larsoft_chno'])
        wfs_good.append(nputils.TH1_to_nparray(f.Get('hWF_GOOD_%0.4d'%larsoft_num)))
        wfs_bad.append(nputils.TH1_to_nparray(f.Get('hWF_BAD_%0.4d'%larsoft_num)))
        plt.plot(np.array(sr),label='(%d,%d,%d)'%(key[0],key[1],key[2]))
    rms_of_rms.append(rms)
#plt.legend()
plt.xlim([30,50])
    #if mean > 5: print key,mean
    
f.Close()
```
