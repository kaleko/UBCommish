```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dc_file = 'zigzag.txt'
dc_array = np.fromfile(dc_file,dtype=float,sep=',')
```

```python
plt.xlabel('Time [tdc]')
plt.ylabel('# of wires')
plt.title('Number of wires that zigzag over a 4-tick window')
plt.grid()

plt.plot(dc_array)
```

```python
#function that takes in dc_array, and looks for pulses (as shown in plot above)
#returns a list of pulse (start_time,width,amplitude)
def find_pulses(wf, min_wires_hit=25, min_pulse_width=4, min_time_bt_pulses=3):

    current_pulse_width, current_pulse_spacing, current_pulse_amp, idx = 0, 0, 0, 0
    pulse_info = []

    #loop over waveform
    for nw in wf:
        if nw > min_wires_hit:
            found_pulse = True
            current_pulse_width += 1
            current_pulse_amp = nw if nw > current_pulse_amp else current_pulse_amp
        else:
            found_pulse = False
            current_pulse_spacing += 1
            
        #Record the pulse info after the pulse is completed, if pulse passes cuts
        if current_pulse_width > min_pulse_width \
            and not found_pulse \
            and current_pulse_spacing > min_time_bt_pulses:
            current_pulse_start = idx-current_pulse_width
            pulse_info.append( (current_pulse_start,current_pulse_width,current_pulse_amp) )
            current_pulse_width, current_pulse_spacing, current_pulse_amp = 0, 0, 0

        idx += 1
    print "Number of pulses above minimum height %d and width %d is %d."%(min_wires_hit,min_pulse_width,len(pulse_info))
    
    return pulse_info
```

```python
big_pulse_info = find_pulses(dc_array)
```

```python
pulse_df = pd.DataFrame(big_pulse_info)
pulse_df.columns = ['start_time','width','amp']
pulse_df = pulse_df.query('amp > 600')
```

```python
#let's overlay all of the big pulses i found
pre_ticks = 0
post_ticks = 0
n_pulses_to_plot = 100000

for index, pulse in pulse_df.iterrows():
    if index >= n_pulses_to_plot: break
    pulse_start_t = pulse['start_time'] - pre_ticks
    if pulse_start_t < 0: continue
    pulse_end_t = pulse['start_time'] + pulse['width'] + post_ticks
    if pulse_end_t > len(dc_array): continue
    y = dc_array[pulse_start_t:pulse_end_t]
    #visualize the pulses where they actually are
    x = np.arange(pulse_start_t,pulse_end_t,1)
    #visualize the pulses all overlaid
    #x = np.arange(0,pulse['width']+pre_ticks+post_ticks,1)
    plt.plot(x,y)
```

```python
#spacings between each consecutive big peak
plt.xlabel('Time between big pulses [tdc]')
plt.ylabel('Counts')
plt.title('Time between Starts of Successive Big Pulses')
plt.grid()
#start at 1 because 0th element is nan
n,bins,patches = plt.hist(np.array(pulse_df['start_time'].diff())[1:],20)
```

```python
#Widths of each big peak
plt.xlabel('Width of big pulses [tdc]')
plt.ylabel('Counts')
plt.title('Width of Individual Big Pulses')
plt.grid()
n,bins,patches = plt.hist(np.array(pulse_df['width']),35)
```
